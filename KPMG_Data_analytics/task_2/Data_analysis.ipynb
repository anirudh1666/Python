This program goes through the excel spreadsheet and calculates age ranges and occupation frequencies.


```python
from openpyxl import load_workbook
import datetime

def calc_age(date):
    
    if date == 'None':
        # Empty cell ignore it
        return -1
    today = datetime.datetime.now()
    today_date = datetime.date(today.year, today.month, today.day)
    dob = datetime.date(int(date[:4]), int(date[5:7]), int(date[8:10]))
    
    return ((today_date - dob) / 365).days


wb = load_workbook('KPMG_Raw_Data.xlsx')
wb.active = 3
sheet = wb.active

ten_to_15 = 0
sixteen_to_20 = 0
twenty1_to_25 = 0
twenty6_to_30 = 0
thirty1_to_35 = 0
thirty6_to_40 = 0
fourty1_to_45 = 0
fourty6_to_50 = 0
fifty1_to_55 = 0
fifty6_to_60 = 0
sixty1_to_65 = 0
sixty6_to_70= 0
seventy_plus = 0

for i in range(1, 1874):
    # i = cur row
    
    age = calc_age(str(sheet.cell(row=i, column=6).value))
    
    if 15 >= age >= 10:
        ten_to_15 += 1
    elif 20 >= age >= 16:
        sixteen_to_20 += 1
    elif 25 >= age >= 21:
        twenty1_to_25 += 1
    elif 30 >= age >= 26:
        twenty6_to_30 += 1
    elif 35 >= age >= 31:
        thirty1_to_35 += 1
    elif 40 >= age >= 36:
        thirty6_to_40 += 1
    elif 45 >= age >= 41:
        fourty1_to_45 += 1
    elif 50 >= age >= 46:
        fourty6_to_50 += 1
    elif 55 >= age >= 51:
        fifty1_to_55 += 1
    elif 60 >= age >= 56:
        fifty6_to_60 += 1
    elif 65 >= age >= 61:
        sixty1_to_65 += 1
    elif 70 >= age >= 66:
        sixty6_to_70 += 1
    elif age > 70:
        seventy_plus += 1

print('10-15: ' + str(ten_to_15))
print('16-20: ' + str(sixteen_to_20))
print('21-25: ' + str(twenty1_to_25))
print('26-30: ' + str(twenty6_to_30))
print('31-35: ' + str(thirty1_to_35))
print('36-40: ' + str(thirty6_to_40))
print('41-45: ' + str(fourty1_to_45))
print('46-50: ' + str(fourty6_to_50))
print('51-55: ' + str(fifty1_to_55))
print('56-60: ' + str(fifty6_to_60))
print('61-65: ' + str(sixty1_to_65))
print('66-70: ' + str(sixty6_to_70))
print('70+: ' + str(seventy_plus))
```

```python
# This part goes through and finds frequency of all occupations

occupations = {}

for i in range(1,1874):
    
    occupation = str(sheet.cell(row=i, column=7).value)
    if occupation in occupations:
        occupations[occupation] += 1
    else:
        occupations[occupation] = 0

for job in occupations:
    print(job + " " + str(occupations[job]))
```
