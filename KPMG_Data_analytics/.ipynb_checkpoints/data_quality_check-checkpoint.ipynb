```python
# This program loops through each sheet and finds all invalid entries.

from openpyxl import load_workbook

wb = load_workbook('KPMG_Raw_Data.xlsx')
```

```python
# First we look at each worksheet in turn and find all empty cell values.
# We want to print out the names of the people with empty cell values 
# and also which cell is empty. 
# First sheet we look at is Transactions

wb.active = 1
cur_sheet = wb.active
total = 0
print('Current sheet:')
print(cur_sheet, end='\n\n')

for i in range(3,20003):
    # i = current row. Cell values are in rows 3 to 20002
    at_row = False         # to stop counting the same row multiple times  
    
    for j in range(1,14):
        # j = current column. There are 14 columns but last column is empty
        # for all values so we ignore it.
        
        if cur_sheet.cell(row=i, column=j).value is None:
            # Empty cell value so print out transaction ID and empty cell.
            print('Transaction ID: ', end=' ')
            print(cur_sheet.cell(row=i, column=1).value)
            print('Empty cell: ', end=' ')
            print(cur_sheet.cell(row=2, column=j).value, end='\n\n')
            
            if not at_row:
                total += 1
                at_row = True

print('Total invalid entries: ' + str(total))
```

```python
# Now we repeat this for Customer Demographic
# Note: We skip the column for default as the entries don't make any sense

wb.active = 3
cur_sheet = wb.active
total = 0
print('Current Sheet: ')
print(cur_sheet, end='\n\n')

for i in range(3,4003):
    # i = current row.
    at_row = False
    
    for j in range(1,14):
        # j = current column.
        
        if j == 11:
            # At column for default. Skip.
            continue
        
        if cur_sheet.cell(row=i, column=j).value is None:
            # Empty cell value so print out customer ID and empty cell.
            print('Customer ID: ', end=' ')
            print(cur_sheet.cell(row=i, column=1).value)
            print('Empty cell: ', end=' ')
            print(cur_sheet.cell(row=2, column=j).value, end='\n\n')
            
            if not at_row:
                total += 1
                at_row = True

print('Total invalid entries: ' + str(total))
        
```

```python
# Finally we repeat for Customer Address

wb.active = 4
cur_sheet = wb.active
total = 0
print('Current Sheet: ')
print(cur_sheet, end='\n\n')

for i in range(3,4002):
    # i = current row.
    at_row = False
    
    for j in range(1,7):
        # j = current column.
        
        if cur_sheet.cell(row=i, column=j).value is None:
            # Empty cell value so print out customer ID and empty cell.
            print('Customer ID: ', end=' ')
            print(cur_sheet.cell(row=i, column=1).value)
            print('Empty cell: ', end=' ')
            print(cur_sheet.cell(row=2, column=j).value, end='\n\n')
            
            if not at_row:
                total += 1
                at_row = True

print('Total invalid entries: ' + str(total))
```

```python
# Now we look at Transactions. We find all transactions older than
# 30 years. We don't want them because they are outdated and don't 
# reflect modern day needs.

import datetime

def in_date_range(date):
    
    if date == 'None':
        # Empty cell ignore it.
        return False
    
    limit = datetime.datetime.now() - datetime.timedelta(days=30*365)
    
    # datetime(year, month, day)
    transaction_date = datetime.date(int(date[:4]), int(date[5:7]), int(date[8:10]))
    limit_date = datetime.date(limit.year, limit.month, limit.day)
    
    if transaction_date < limit_date:
        return True
    
    return False
    
    

wb.active = 1
cur_sheet = wb.active
total = 0
print('Current sheet: ')
print(cur_sheet, end='\n\n')

for i in range(3,20003):
    # look at each row. Calculate their age and if its older than 30 years list them.
    
    if in_date_range(str(cur_sheet.cell(row=i, column=4).value)):
        total += 1
        print('Transaction ID: ', end=' ')
        print(cur_sheet.cell(row=i, column=1).value)

print('Total invalid entries: ' + str(total))
```

```python
# Now we repeat this for Customer demographic but with max age limit of 110

import datetime

def in_age_range(date):
    
    if date == 'None':
        # Empty cell value we ignore it.
        return False
    
    limit = datetime.datetime.now() - datetime.timedelta(days=110*365)
    
    birth_date = datetime.date(int(date[:4]), int(date[5:7]), int(date[8:10]))
    limit_date = datetime.date(limit.year, limit.month, limit.day)
    
    if birth_date < limit_date:
        return True
    
    return False


wb.active = 3
cur_sheet = wb.active
total = 0
print('Current sheet: ')
print(cur_sheet, end='\n\n')

for i in range(3,4003):
    
    if in_age_range(str(cur_sheet.cell(row=i, column=6).value)):
        total += 1
        print('Customer ID: ', end=' ')
        print(cur_sheet.cell(row=i, column=1).value)
        
print('Total invalid entries: ' + str(total))
```
